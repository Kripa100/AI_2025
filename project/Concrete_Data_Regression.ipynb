```python
df_copy = df1.copy()

for column in df_copy.columns:
  #IQR method for handling outliers
  Q1 = df_copy[column].quantile(0.25)
  Q3 = df_copy[column].quantile(0.75)
  IQR = Q3 - Q1
  #defining lower and upper bounds
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR
  #replacing outliers with mean values
  df_copy.loc[df_copy[column] < lower_bound, column] = df_copy[column].mean().astype(int)
  df_copy.loc[df_copy[column] > upper_bound, column] = df_copy[column].mean().astype(int)
```

```python
#boxplots for each column after outlier handling

df_copy.plot(kind='box', subplots=True, layout=(3,3), sharex=False, sharey=False)
plt.suptitle("Box Plots of Features after outlier handling")
plt.tight_layout()
plt.show()
```

```python
#Load our X features and y label
X = df_copy.drop('concrete_compressive_strength',axis=1).values
y = df_copy['concrete_compressive_strength'].values

#Split our data as training data and testing data
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=17)
```

```python
#Scaling the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
y_train_scaled = scaler.fit_transform(y_train.reshape(-1, 1)).flatten()
y_test_scaled = scaler.transform(y_test.reshape(-1, 1)).flatten()
```

```python
df_scaled = pd.DataFrame(X_train_scaled, columns=df_copy.drop('concrete_compressive_strength', axis=1).columns)

#Box-plot after Scaling
df_scaled.plot(kind='box', subplots=True, layout=(3,3), sharex=False, sharey=False, figsize=(12, 10))
plt.suptitle("Box Plots of Scaled Features")
plt.tight_layout()
plt.show()
```

```python
#Save df_scaled as csv file

df_scaled.to_csv('df_scaled.csv', index=False)
```

```python
#Creating a distplot for all numerical features

# Create a figure and a grid of subplots
fig, axes = plt.subplots(3, 3, figsize=(15, 15))

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Iterate through the columns and the axes
for i, column in enumerate(df_scaled.columns):
    # Plot the histogram with KDE using seaborn.histplot
    sns.histplot(data=df_scaled, x=column, kde=True, ax=axes[i])
    axes[i].set_title(column)


plt.tight_layout() # Adjust layout to prevent overlapping titles and labels
plt.suptitle("Histograms with KDE of Scaled Features", y=1.02)
plt.show()
```

```python
models = {
    "Linear Regression": LinearRegression(),
    "Ridge": Ridge(alpha=0.1),
    "Lasso": Lasso(alpha=0.1),
    "dsr": DecisionTreeRegressor(max_depth=5,min_samples_split=5),
    "rfr": RandomForestRegressor(n_estimators=50, max_depth=10, min_samples_split=5),
    "svr": SVR()
    }

#Creating a list that will hold cross-validation scores of all models
results = []

#Creating a KFold object
#This performs training and validation for a same model for 7 times
#Every time, the validation data changes.
kf = KFold(n_splits=7, random_state=42, shuffle=True)

# Loop through the models' values
for model in models.values():

  # Perform cross-validation
  cv_scores = cross_val_score(model, X_train_scaled, y_train_scaled, cv=kf)

  # Append the results
  results.append(cv_scores)

# Create a box plot of the results
plt.boxplot(results, tick_labels=models.keys())
plt.title("Cross-Validation Results")
plt.xlabel("Model")
plt.ylabel("Mean Squared Error")
plt.show()
```

```python
test_mse = {}
for name, model in models.items():

  # Fit the model to the training data
  model.fit(X_train_scaled,y_train_scaled)

  # Make predictions on the test set
  y_pred = model.predict(X_test_scaled)

  # Calculate the test_rmse
  test_mse[name] = mean_squared_error(y_test_scaled, y_pred)
  print(f"{name} Test Set MSE: {test_mse[name]}")
```

```python
#bar graph to show mse of all models

sns.barplot(x=list(models.keys()), y=test_mse.values())
plt.title("Mean Square Error of all models")
plt.xlabel("Model")
plt.ylabel("Mean Squared Error")
plt.show()
```

```python
#Create the model
regression_model_baseline = RandomForestRegressor(random_state=42)

# Fit to the training data
regression_model_baseline.fit(X_train_scaled, y_train_scaled)
```

```python
#Make predictions on the test data using the tuned model
y_pred_scaled = regression_model_baseline.predict(X_test_scaled)

#Visualize the model prediction using scatter-plot
plt.scatter(y_test_scaled, y_pred_scaled)
plt.xlabel("Actual Scaled Concrete compressive strength")
plt.ylabel("Predicted Scaled Concrete compressive strength")
plt.plot(y_test_scaled,y_test_scaled, color='red')
plt.title("Actual vs Predicted Scaled Concrete compressive strength (Baseline Model)")
plt.show()
```

```python
#Squared errors
mse = mean_squared_error(y_test_scaled, y_pred_scaled)
rmse = np.sqrt(mse) # Calculate RMSE
r2 = r2_score(y_test_scaled, y_pred_scaled)

print(f"Random Forest Regressor Test Set MSE: {mse}")
print(f"Random Forest Regressor Test Set RMSE: {rmse}")
print(f"Random Forest Regressor Test Set R-squared: {r2}")
```

```python
#Absolute errors
mae = mean_absolute_error(y_test_scaled, y_pred_scaled)
print(f"Mean Absolute Error: {mae}")
```

```python
# Printing the metrics on original data using the baseline model
y_pred_scaled_baseline = regression_model_baseline.predict(X_test_scaled)

print("Baseline Model\n")

# Inverse transform the predictions back to the original scale
y_pred_unscaled_baseline = scaler.inverse_transform(y_pred_scaled_baseline.reshape(-1, 1)).flatten()

# Squared errors on unscaled data
mse_unscaled_baseline = mean_squared_error(y_test, y_pred_unscaled_baseline)
rmse_unscaled_baseline = np.sqrt(mse_unscaled_baseline)
r2_unscaled_baseline = r2_score(y_test, y_pred_unscaled_baseline)
print(f"Baseline Random Forest Regressor Test Set MSE (Unscaled): {mse_unscaled_baseline}")
print(f"Baseline Random Forest Regressor Test Set RMSE (Unscaled): {rmse_unscaled_baseline}")
print(f"Baseline Random Forest Regressor Test Set R-squared (Unscaled): {r2_unscaled_baseline}")

# Absolute errors on unscaled data
mae_unscaled_baseline = mean_absolute_error(y_test, y_pred_unscaled_baseline)
print(f"Mean Absolute Error (Unscaled): {mae_unscaled_baseline}")

#Log errors on unscaled data
msle_unscaled_baseline = mean_squared_log_error(y_test, y_pred_unscaled_baseline)
rmsle_unscaled_baseline = np.sqrt(msle_unscaled_baseline)
print(f"Baseline Random Forest Regressor Test Set MSLE (Unscaled): {msle_unscaled_baseline}")
print(f"Baseline Random Forest Regressor Test Set RMSLE (Unscaled): {rmsle_unscaled_baseline}")
```

```python
#Visualize the model prediction using scatter-plot
plt.scatter(y_test, y_pred_unscaled_baseline)
plt.xlabel("Actual UnScaled Concrete compressive strength")
plt.ylabel("Predicted UnScaled Concrete compressive strength")
plt.plot(y_test,y_test, color='red')
plt.title("Actual vs Predicted UnScaled Concrete compressive strength (Baseline Model)")
plt.show()
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
param_grid = {
                  "n_estimators":[50,100,150,200, 250, 300],
                  "max_depth":[None, 10, 20, 30, 40],
                  "min_samples_split":[2,4,6, 8, 10],
                  "min_samples_leaf":[1, 2, 3],
                  "bootstrap":[True, False]
            }

rfr_cv = RandomizedSearchCV(regression_model_baseline, param_distributions=param_grid, cv=kf, n_iter=50, random_state=42)
```

```python
# Fit the RandomizedSearchCV object to the scaled training data
rfr_cv.fit(X_train_scaled, y_train_scaled)
```

```python
#Get the tuned model
regression_model_tuned = rfr_cv.best_estimator_
```

```python
#Make predictions on the test data using the tuned model
y_pred_tuned = regression_model_tuned.predict(X_test_scaled)

#Visualize the model prediction using scatter-plot
plt.scatter(y_test_scaled, y_pred_tuned)
plt.xlabel("Actual Scaled Concrete compressive strength")
plt.ylabel("Predicted Scaled Concrete compressive strength")
plt.plot(y_test_scaled,y_test_scaled, color='red')
plt.title("Actual vs Predicted Scaled Concrete compressive strength(Tuned Model)")
plt.show()
```

```python
#squared errors
mse_tuned = mean_squared_error(y_test_scaled, y_pred_tuned)
rmse_tuned = np.sqrt(mse_tuned) # Calculate RMSE
r2_tuned = r2_score(y_test_scaled, y_pred_tuned)

print(f"Tuned Random Forest Regressor Test Set MSE: {mse_tuned}")
print(f"Tuned Random Forest Regressor Test Set RMSE: {rmse_tuned}")
print(f"Tuned Random Forest Regressor Test Set R-squared: {r2_tuned}")
```

```python
#Absolute errors
from sklearn.metrics import mean_absolute_error
mae_tuned = mean_absolute_error(y_test_scaled, y_pred_tuned)

print(f"Mean Absolute Error: {mae_tuned}")
```

```python
print("Tuned model\n")

# Printing the metrics on original data using tuned model

y_pred_scaled_tuned = regression_model_tuned.predict(X_test_scaled)

# Inverse transform the predictions back to the original scale
y_pred_unscaled_tuned = scaler.inverse_transform(y_pred_scaled_tuned.reshape(-1, 1)).flatten()

# Squared errors on unscaled data
mse_unscaled_tuned = mean_squared_error(y_test, y_pred_unscaled_tuned)
rmse_unscaled_tuned = np.sqrt(mse_unscaled_tuned)
r2_unscaled_tuned = r2_score(y_test, y_pred_unscaled_tuned) # Calculate R-squared on unscaled data

print(f"Tuned Random Forest Regressor Test Set MSE (Unscaled): {mse_unscaled_tuned}")
print(f"Tuned Random Forest Regressor Test Set RMSE (Unscaled): {rmse_unscaled_tuned}")
print(f"Tuned Random Forest Regressor Test Set R-squared (Unscaled): {r2_unscaled_tuned}")

# Absolute errors on unscaled data
mae_unscaled_tuned = mean_absolute_error(y_test, y_pred_unscaled_tuned)

print(f"Mean Absolute Error (Unscaled): {mae_unscaled_tuned}")

# Log errors on unscaled data
# Note: MSLE and RMSLE are typically used for positive target values.
# Ensure y_test and y_pred_unscaled_tuned are positive before calculating.
# In this dataset, concrete strength is always positive, so it's fine.
msle_unscaled_tuned = mean_squared_log_error(y_test, y_pred_unscaled_tuned)
rmsle_unscaled_tuned = np.sqrt(msle_unscaled_tuned)

print(f"Tuned Random Forest Regressor Test Set MSLE (Unscaled): {msle_unscaled_tuned}")
print(f"Tuned Random Forest Regressor Test Set RMSLE (Unscaled): {rmsle_unscaled_tuned}")
```

```python
#Visualize the model prediction using scatter-plot
plt.scatter(y_test, y_pred_unscaled_tuned)
plt.xlabel("Actual UnScaled Concrete compressive strength")
plt.ylabel("Predicted UnScaled Concrete compressive strength")
plt.plot(y_test,y_test, color='red')
plt.title("Actual vs Predicted UnScaled Concrete compressive strength (Tuned Model)")
plt.show()
```
